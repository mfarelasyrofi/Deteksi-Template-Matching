Mencocokkan Citra menggunakan Metode Template Matching

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img1 = cv2.imread('tiger2.jpg')
img1 = cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)
img2 = cv2.imread('tiger.jpg')
img2 = cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
method = ['cv2.TM_CCOEFF']
for m in method:
  img_copy1 = img1.copy()
  mymethod = eval(m)
  res = cv2.matchTemplate(img_copy1,img2,mymethod)
  min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
  if mymethod in [cv2.TM_SQDIFF, cv2.TM_SQDIFF_NORMED]:
    top_left = min_loc
  else:
      top_left = max_loc
  h,w,c = img1.shape

  bottom_right = (top_left[0]+w,top_left[1]+h)
  cv2.rectangle(img_copy1,top_left,bottom_right,(255,0,0),5)

  plt.subplot(1,2,1)
  plt.imshow(res)
  plt.title('HEATMAP OF TEMPLATE MATCHING')

  plt.subplot(1,2,2)
  plt.imshow(img_copy1)
  plt.title('DETECTION OF TEMPLATE')
```
